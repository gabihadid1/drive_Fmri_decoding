```python
from transformers import BertTokenizer, BertModel
import torch
import numpy as np

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")
model = BertModel.from_pretrained("bert-base-uncased").to(device)
model.eval()

def get_sentence_embedding(sentence):
    inputs = tokenizer(sentence, return_tensors="pt", truncation=True, max_length=128).to(device)
    with torch.no_grad():
        outputs = model(**inputs)
    token_embeddings = outputs.last_hidden_state[0][1:-1]
    return token_embeddings.mean(dim=0).cpu().numpy()

```

```python
from learn_decoder import learn_decoder

def cosine_similarity(u, v):
    return np.dot(u, v) / (np.linalg.norm(u) * np.linalg.norm(v))

def rank_accuracy(decoded, matrix_semantics, true_index):
    acc = {i: cosine_similarity(decoded, matrix_semantics[i, :]) for i in range(matrix_semantics.shape[0])}
    sorted_indices = [idx for idx, _ in sorted(acc.items(), key=lambda kv: kv[1], reverse=True)]
    return sorted_indices.index(true_index) + 1

def cross_validate(data_matrix, semantic_matrix,n_sentence = 384):
    n_folds = 18
    size = n_sentence // n_folds
    all_fold_ranks = []
    average_ranks = []

    for i in range(n_folds):
        start = i * size
        end = start + size
        test_indices = list(range(start, end))
        train_indices = [j for j in range(n_sentence) if j not in test_indices]

        M = learn_decoder(data_matrix[train_indices], semantic_matrix[train_indices])
        for j in test_indices:
            decoded = data_matrix[j] @ M
            rank = rank_accuracy(decoded, semantic_matrix, j)
            all_fold_ranks.append((j, rank))
        average_ranks.append(np.mean([r for (_, r) in all_fold_ranks[-size:]]))

    return all_fold_ranks, average_ranks
```

```python
import random
import spacy
from typing import List, Dict
import math

nlp = spacy.load("en_core_web_sm")

def is_content_word(token):
    return (not token.is_stop) and (not token.is_punct) and (token.pos_ != "SPACE")

def generate_fragments(sentences: List[str]) -> Dict[int, Dict[str, str]]:
    random.seed(42)

    """
    sentences: List of full sentence strings
    Returns: Dictionary mapping sentence index to dicts of fragment types
    """
    all_fragments = {}

    for i, sentence in enumerate(sentences):
        doc = nlp(sentence)

        # Clean word list: no punctuation, no space-only tokens
        words = [token.text.lower() for token in doc if not token.is_punct and not token.is_space]
        content_words = [token.text.lower() for token in doc if is_content_word(token)]


        L = len(words)
        C = len(content_words)

        fragments = {
            "full_clean": " ".join(words), # Cleaned sentence (lowercased, no punctuation)
            "content": " ".join(content_words) # Only content words (lowercased, no stopwords/punct)
        }

        for pct in [0.25, 0.5, 0.75]:
            # Compute number of tokens to use
            n_words = max(1, math.ceil(L * pct))
            n_content = max(1, math.ceil(C * pct)) # maybe change to n_words = min(L - 1, max(1, math.ceil(L * pct)))

            # standard word-based fragments
            fragments[f"first_{int(pct*100)}"] = " ".join(words[:n_words])
            fragments[f"last_{int(pct*100)}"] = " ".join(words[-n_words:])
            fragments[f"random_{int(pct*100)}"] = " ".join(random.sample(words, min(n_words, L)))

            # content word-based fragments
            fragments[f"first_{int(pct*100)}_content"] = " ".join(content_words[:n_content])
            fragments[f"last_{int(pct*100)}_content"] = " ".join(content_words[-n_content:])
            fragments[f"random_{int(pct*100)}_content"] = " ".join(random.sample(content_words, min(n_content, C)))


        all_fragments[i] = fragments

    return all_fragments
```

```python

def load_glove_full(filepath):
    glove = {}
    with open(filepath, "r", encoding="utf8") as f:
        for line in f:
            parts = line.strip().split()
            word = parts[0]
            try:
                vector = np.array(parts[1:], dtype=np.float32)
                if vector.shape[0] == 300:
                    glove[word] = vector
            except ValueError:
                continue  # skip malformed lines
    return glove

# Load full GloVe
glove = load_glove_full("glove.42B.300d.txt")


```

```python
import string

def get_glove_embedding(text, glove, dim=300):
    doc = nlp(text.lower())
    vectors = []

    for token in doc:
        if token.is_punct or token.is_space:
            continue
        lemma = token.lemma_.lower().strip(string.punctuation)
        if lemma in glove:
            vectors.append(glove[lemma])

    if not vectors:
        return np.zeros(dim)
    return np.mean(vectors, axis=0)
```

```python
import random
import spacy
from typing import List, Dict
import math

nlp = spacy.load("en_core_web_sm")

def is_content_word(token):
    return (not token.is_stop) and (not token.is_punct) and (token.pos_ != "SPACE")

def generate_fragments(sentences: List[str]) -> Dict[int, Dict[str, str]]:
    random.seed(42)

    """
    sentences: List of full sentence strings
    Returns: Dictionary mapping sentence index to dicts of fragment types
    """
    all_fragments = {}

    for i, sentence in enumerate(sentences):
        doc = nlp(sentence)

        # Clean word list: no punctuation, no space-only tokens
        words = [token.text.lower() for token in doc if not token.is_punct and not token.is_space]
        content_words = [token.text.lower() for token in doc if is_content_word(token)]


        L = len(words)
        C = len(content_words)

        fragments = {
            "full_clean": " ".join(words), # Cleaned sentence (lowercased, no punctuation)
            "content": " ".join(content_words) # Only content words (lowercased, no stopwords/punct)
        }

        for pct in [0.25, 0.5, 0.75]:
            # Compute number of tokens to use
            n_words = max(1, math.ceil(L * pct))
            n_content = max(1, math.ceil(C * pct)) # maybe change to n_words = min(L - 1, max(1, math.ceil(L * pct)))

            # standard word-based fragments
            fragments[f"first_{int(pct*100)}"] = " ".join(words[:n_words])
            fragments[f"last_{int(pct*100)}"] = " ".join(words[-n_words:])
            fragments[f"random_{int(pct*100)}"] = " ".join(random.sample(words, min(n_words, L)))

            # content word-based fragments
            fragments[f"first_{int(pct*100)}_content"] = " ".join(content_words[:n_content])
            fragments[f"last_{int(pct*100)}_content"] = " ".join(content_words[-n_content:])
            fragments[f"random_{int(pct*100)}_content"] = " ".join(random.sample(content_words, min(n_content, C)))


        all_fragments[i] = fragments

    return all_fragments
```

```python
def build_fragment_matrix(embed_dict, frag_type):
    X = []
    idx_map = []
    for idx in sorted(embed_dict.keys()):
        emb = embed_dict[idx].get(frag_type)
        if emb is not None and not np.any(np.isnan(emb)):
            X.append(emb)
            idx_map.append(idx)
    return np.vstack(X), idx_map  # shape: (n_sentences, embedding_dim)
def predict_fmri(X, model):
    return X @ model.coef_.T + model.intercept_  # shape: (n_sentences, n_voxels)
```

```python
from scipy.stats import pearsonr

def compute_pearson(y_true, y_pred):

    r_vals = []
    for v in range(y_true.shape[1]):
        r, _ = pearsonr(y_true[:, v], y_pred[:, v])
        r_vals.append(r)
    return np.array(r_vals)  # shape: (n_voxels,)
```

```python
def get_roi_scores(r_vals, roi_df):
    df = roi_df.copy()
    df['r'] = r_vals
    return df.groupby("roi")["r"].mean().sort_values(ascending=False)
```

```python
from sklearn.linear_model import RidgeCV
from tqdm import tqdm

results_by_fragment_bert = {}

for frag in ['first_25', 'first_50', 'first_75','last_25', 'last_50','last_75','full_clean','random_25','random_50', 'random_75']:
    print(f"Processing fragment: {frag}")

    # Filter and align data
    valid_pairs = [
        (idx, embed_dict_bert[idx].get(frag))
        for idx in sorted(embed_dict_bert)
        if frag in embed_dict_bert[idx]
        and embed_dict_bert[idx][frag] is not None
        and not np.any(np.isnan(embed_dict_bert[idx][frag]))
        and sentence_map.get(idx) in fmri_avg
    ]
    X_clean = []
    y_clean = []
    for idx, emb in valid_pairs:
        sentence = sentence_map[idx]
        if sentence in roi_scalars.index:
            roi_vec = roi_scalars.loc[sentence].values
            if all(isinstance(v, np.ndarray) and not np.any(np.isnan(v)) for v in roi_vec):
                X_clean.append(emb)
                y_clean.append(np.concatenate(roi_vec))


    X_bert = np.vstack(X_clean)
    y_bert = np.vstack(y_clean)



    # Train model (multi-output Ridge regression)
    ridge = RidgeCV(alphas=[0.1, 1.0, 10.0])
    ridge.fit(X_bert, y_bert)

    # Predict
    y_pred = ridge.predict(X_bert)

    # Pearson per voxel
    r_vals = compute_pearson(y_bert, y_pred)

    # Aggregate by ROI
    roi_r = get_roi_scores(r_vals, roi_df)
    results_by_fragment_bert[frag] = roi_r
```
